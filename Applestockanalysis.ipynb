```python
#By using yfinance day to day stock values can be used
import yfinance as yf
import pandas as pd
```

```python
#considering stocks of apple,for understanding I have taken current years data
ticker='AAPL'
data=yf.download(ticker,period='ytd')
```

```python
#printing the head of the data
print("Head of the data")
data.head()
```

```python
#printing tail of the data
print("Tail of the data")
data.tail()
```

```python
#printing infromation such as missing values,number of rows,colums etc from the data
print(data.info())
```

```python
#basic statistics of each colum
print(data.describe())
```

```python
#checking null values
print(data.isnull().sum())
```

```python
#visulaizing the data
#Opening prize 
import matplotlib.pyplot as plt
sp=data['Open'].plot(figsize=(14,7),title='Opening price of Apple Stock',grid=True)
sp.set_xlabel('Start date')
sp.set_ylabel('Price(USD)')
plt.show()
```

```python
#Visualizing the close prize 
ep=data['Close'].plot(figsize=(14,7),title='Closing price of Apple Stock',grid=True)
ep.set_xlabel("Date")
ep.set_ylabel("Price(USD)")
plt.show()
```

```python
#analyzing the volume traded
v=data['Volume'].plot(kind='bar',figsize=(14,7),grid=True,title='Volume traded')
v.set_xlabel("Date")
v.set_ylabel("Quantity")
plt.show()
```

```python
#analyzing percentage change in daily returns
data['Daily return']=data['Close'].pct_change()
dr=data['Daily return'].plot(kind='hist',bins=50,title='Daily returns')
plt.show()
```

```python
#Correlation analysis - undestanding relationship between differt columns
import seaborn as sns
cm=data.corr()
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,cmap='coolwarm')
plt.title("Correlation matrix")
plt.show()
```

```python
#identifying outliers
plt.figure(figsize=(10,7))
sns.boxplot(x=data['Close'])
plt.show()
```

```python
#monthly resampling
monthly_data=data['Close'].resample('M').mean()
monthly_data.plot(kind='bar',color='red',title='Monthly Resampling',xlabel='Month',ylabel='Prize')
plt.show()
```

```python
#weekly resampling
weekly_data=data['Close'].resample('W').mean()
weekly_data.plot(kind='bar',color='green',title='Weekly resampling',xlabel='Week',ylabel='Prize')
plt.show()
```
